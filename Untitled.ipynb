```python
import nltk
```

```python
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
nltk.download('stopwords')
nltk.download('punkt_tab')
stop_words = set(stopwords.words('english'))
def clean_text(text):
    #1. lower case
    text = text.lower()
    #2. Remove Special Symbols
    text = re.sub(r'[^a-zA-Z\s]','',text)
    #3. Remove HTTP links
    text = re.sub(r'https\S','',text)
    #4. Word Tokenize
    tokens = word_tokenize(text)
    #5. removing stop words
    # l = []
    # for word in tokens:
    #     if word not in stop_words:
    #         l.append(word)
    sw_text = [word for word in tokens if word not in stop_words]
    #6. Applying stemming or lemmatization
    # l = []
    # for word in sw_text:
    #     l.append(stemmer.stem(word))
    clean_text = [stemmer.stem(word) for word in sw_text]
    
    return " ".join(clean_text)

text = "Hello this is NLP that help @@##$$%$^$$#$#  98374528973498 https://coderameen.com is a that great "
clean_text(text)
```

```python
import pandas as pd
df = pd.read_csv("SMS fraud Dataset.csv")
df.head()
```

```python
df['clean_sms'] = df['sms'].apply(clean_text)
```

```python
#To check if my dataset has null or missing values
df.isnull().sum()
```

```python
#split into X and y(input and ouput)
X = df['clean_sms']#input
y = df['label']

df['label'].value_counts()
```

```python
from imblearn.under_sampling import RandomUnderSampler
#from imblearn.over_sampling import RandomOverSampler
#from imblearn.over_sampling import SMOTE
rus = RandomUnderSampler()
X = df['clean_sms'].values.reshape(-1,1)
y = df['label']
X_balance,y_balance = rus.fit_resample(X,y)
df = pd.DataFrame({'clean_sms':X_balance.flatten(),'label':y_balance})
```

```python
df['label'].value_counts()
```

```python
df.head()
```

```python
#Train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df['clean_sms'],df['label'], test_size=0.2, random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train_tfidf,y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
y_pred = model.predict(X_test_tfidf)
y_pred
```

```python
acc = accuracy_score(y_test,y_pred)
acc * 100
```

```python
#confusion matric
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
#classification report 
cr = classification_report(y_test,y_pred)
cr
```

```python
import pickle
#save the model and tfidf
model = pickle.dump(model,open('model.pkl','wb'))
tfidf = pickle.dump(tfidf, open('tfidf.pkl','wb'))
```

```python
#To pickle pickle file
model  = pickle.load(open('model.pkl','rb'))
tfidf = pickle.load(open('tfidf.pkl','rb'))
```

```python
def fraud_detection(sms):
    sms = clean_text(sms)
    #convert text to number
    features = tfidf.transform([sms])
    pred = model.predict(features)
    if pred.item() == 1:
        return  "Fraud SMS"
    else:
        return "Real SMS"
# sms = "Free Entry ticket for going paris and dubai"
sms = "can we have meeting today"
fraud_detection(sms)
```
